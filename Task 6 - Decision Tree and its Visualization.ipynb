### Task 6: Create the Decision Tree classifier and visualize it graphically.(Iris dataset)

* **Problem Statement - we classify obeservation using decision tree and show graphically how classification is actually done.**

```python
# Importing libraries in Python
import sklearn.datasets as datasets
import pandas as pd

# Loading the iris dataset
iris=datasets.load_iris()

# Forming the iris dataframe
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head(5))

y=iris.target
print(y)
```

## Data Preparation
Here we divide data into Features(independent variable(s)) and Label (dependent variable).  
After That We split data into 80% for Traing the model and 20% for Testing that model.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df[iris.feature_names], y, random_state=0)
```

## Training Model
Now we will use training data to train Decision Tree model.

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(criterion="entropy",
                             random_state = 0)
# Step 3: Train the model on the data
clf.fit(X_train,Y_train)
# Step 4: Predict labels of unseen (test) data
# Not doing this step in the tutorial
y_pred=clf.predict(X_test)
```

### Visualization of Decision Tree

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(clf);
```

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(clf,
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('imagename.png')
```

### Accuracy of model

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(Y_test, y_pred, normalize=True, sample_weight=None)
```
